```python
a=input( )


if len(a) != 1:
  print("Please enter a single character.")
else:
  char_val = ord(a)

  if (char_val >= 65 and char_val <= 90):
    print('uppercase')
  elif (char_val >= 97 and char_val <= 122):
    print('lowercase')
  elif (char_val >= 48 and char_val <= 57):
    print('digit')
  else:
    print('other character')
```
